````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# Load your dataset into a Pandas DataFrame (replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv('/content/train.csv')  # Or pd.read_excel('your_dataset.xlsx') if it's an Excel file

# Select the features you want to reduce dimensionality on -  using actual column names from your DataFrame
X = df[['age', 'result', 'A1_Score', 'A2_Score', 'A3_Score', 'A4_Score', 'A5_Score', 'A6_Score', 'A7_Score', 'A8_Score', 'A9_Score', 'A10_Score', 'result']]

# Choose the number of components you want to keep
n_components = 2  # You can adjust this based on the desired level of dimensionality reduction

# Create a PCA object with the specified number of components
pca = PCA(n_components=n_components)

# Fit the PCA model to your data
pca.fit(X)

# Transform your data into the reduced-dimensional space
X_reduced = pca.transform(X)

# Create a new DataFrame with the reduced dimensions
df_reduced = pd.DataFrame(X_reduced, columns=['PC1', 'PC2'])

# You can now use df_reduced for further analysis or visualization
print(df_reduced.head())

# Visualize the reduced data (for example, with a scatter plot)
plt.figure(figsize=(8, 6))
plt.scatter(df_reduced['PC1'], df_reduced['PC2'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Reduced Data')
plt.show()

# Explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance_ratio)

# Cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print("Cumulative Explained Variance:", cumulative_explained_variance)
````
